### Recheneffizienz

**`zip`**:

```python
for element in [1, 2, 3, 4]:
    print(element)



l_1 = [1, 2, 3, 4]
l_2 = [200, 300, 400, 500]

print()

for i in range(len(l_1)):
    print( l_1[i] + l_2[i] )
    

print()
    
# zip


for el_1, el_2 in zip(l_1, l_2):
    
    print(el_1, el_2)
```

```python
# listen elementweise verrechnen

def add_element_wise(l_1, l_2):
    
    added_list = []
    
    for el_1, el_2 in zip(l_1, l_2):
        
        added_list.append(el_1 + el_2)
        
    return added_list

add_element_wise(l_1, l_2)
```

```python
l_1 = []
l_2 = []

for i in range(10000000):
    
    l_1.append(i)
    l_2.append((i + 2) * 100)
    
add_element_wise(l_1, l_2)
```

```python
a_1 = np.array(l_1)
a_2 = np.array(l_2)
```

```python
a_1 + a_2
```

```python
%timeit add_element_wise(l_1, l_2)
```

```python
%timeit a_1 + a_2
```
