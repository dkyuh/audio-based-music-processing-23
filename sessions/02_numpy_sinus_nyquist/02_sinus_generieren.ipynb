```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
```

# Sinus generieren

```python
print(np.round(np.sin(0), 3))

print(np.round(np.sin(1 * np.pi / 4), 3))
print(np.round(np.sin(1 * np.pi / 2), 3))
print(np.round(np.sin(3 * np.pi / 4), 3))

print(np.round(np.sin(1 * np.pi), 3))

print(np.round(np.sin(5 * np.pi / 4), 3))
print(np.round(np.sin(3 * np.pi / 2), 3))
print(np.round(np.sin(7 * np.pi / 4), 3))

print(np.round(np.sin(2 * np.pi), 3))
```

```python
sr = 50 # sampling range / sample rate
freq = 3 # in Hz
amp = 0.5
t = np.arange(sr) / sr * 2 * np.pi

x = amp * np.sin( t * freq )

plt.plot(x, '.')
plt.ylim(-1, 1)
plt.show()
```

```python
sr = 20000 # sampling rate / sample rate

freq = 50 # in Hz
amp = 0.75

t = np.arange(sr) / sr * 2 * np.pi

x = amp * np.sin(t * freq)

plt.figure(figsize=(12, 4))
plt.plot(x, '-')
plt.ylim(-1, 1)
plt.show()

display(Audio(x, rate=sr, normalize=False))
```
