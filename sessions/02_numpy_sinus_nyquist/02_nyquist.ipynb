# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
```

# Nyquist-Grenze

```python
for freq in range(0, 100):

    sr = 100
    amp = 0.5
    t = np.arange(sr) / sr * 2 * np.pi

    x = amp * np.sin( t * freq )

    plt.figure(figsize=(12, 4))
    plt.plot(x, '-')
    plt.ylim(-1, 1)
    plt.show()
```
